```python
import pandas as pd
```

## Data Collection of COVID 10K Dataset

```python
encounters = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\encounters.csv'
patients = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\patients.csv'
conditions = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\conditions.csv'
mediations = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\medications.csv'
immunizations = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\immunizations.csv'
careplans = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\careplans.csv'
procedures = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\procedures.csv'
observations = '10k_synthea_covid19_csv\10k_synthea_covid19_csv\observations.csv'
```

```python
# Let's read the encounters file
encounters_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\encounters.csv')
```

```python
encounters_df.head()
```

```python
# Let's remove the columns that we don't need
encounters_df = encounters_df.drop(columns=['ORGANIZATION', 'PROVIDER', 'PAYER', 'ENCOUNTERCLASS', 'CODE', 'BASE_ENCOUNTER_COST', 'TOTAL_CLAIM_COST', 'PAYER_COVERAGE', 'REASONCODE'])

# Change Description to Encounter_Description
encounters_df = encounters_df.rename(columns={'DESCRIPTION': 'ENCOUNTER_DESCRIPTION'})

encounters_df.head()
```

```python
# Number of unique enounters
print('The total number of unique encounters is: ', encounters_df['Id'].nunique())
```

```python
# Let's read the patients file
patients_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\patients.csv')
```

```python
patients_df.head()
```

```python
# Let's remove the columns we don't need
patients_df = patients_df.drop(columns=['SSN', 'DRIVERS', 'PASSPORT', 'PREFIX', 'FIRST', 'LAST', 'SUFFIX', 'MAIDEN', 'BIRTHPLACE', 'ADDRESS', 'CITY', 'STATE', 'COUNTY', 'ZIP', 'LAT', 'LON', 'HEALTHCARE_EXPENSES', 'HEALTHCARE_COVERAGE'])

#change ID column to PATIENT
patients_df.rename(columns={'Id':'PATIENT'}, inplace=True)
patients_df.head()
```

```python
# Value count for unique patients
unique_patients = patients_df['PATIENT'].nunique()
print(f'The number of unique patients is {unique_patients}')
```

```python
# Let's join the encounters and patients files at the Id column for patients and PATIENT column for encounters
merged_df = pd.merge(encounters_df, patients_df, on='PATIENT')
# Change Id column to ENCOUNTER
merged_df.rename(columns={'Id':'ENCOUNTER'}, inplace=True)
merged_df.head()
```

```python
# Let's read the observations file

observations_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\observations.csv')
```

```python
observations_df.head()
```

```python
# Let's remove the columns we don't need
observations_df = observations_df.drop(columns=['DATE', 'CODE', 'UNITS', 'TYPE'])

observations_df.head()
```

```python
top_observations = observations_df['DESCRIPTION'].value_counts()
top_observations.head(10)
```

```python
# Let's keep only Diastolic Blood Pressure, Systolic Blood Pressure, Body Weight, Heart Rate, Respiratory Rate, Oxygen Saturation in Arterial Blood

observations_df = observations_df[observations_df['DESCRIPTION'].isin(['Diastolic Blood Pressure', 'Systolic Blood Pressure', 'Body Weight', 'Heart rate', 'Respiratory rate', 'Oxygen saturation in Arterial blood'])]

# Value counts for Description
print(observations_df['DESCRIPTION'].value_counts())
```

```python
#Let's create a pivot table with the desciption as columns and values as the values

pivot_table = observations_df.pivot_table(index=['PATIENT', 'ENCOUNTER'], columns='DESCRIPTION', values='VALUE', aggfunc='first').reset_index()
```

```python
pivot_table.head()
```

```python
# Remove rows with missing values
pivot_table = pivot_table.dropna()

pivot_table.info()
```

```python
# Order the pivot table by date and patient
pivot_table = pivot_table.sort_values(by=['PATIENT', 'ENCOUNTER'])
```

```python
# Merge the pivot table with the merged_df

final_df = pd.merge(merged_df, pivot_table, on=['PATIENT', 'ENCOUNTER'])

final_df.head()
```

```python
# Number of unique encounters
unique_encounters = final_df['ENCOUNTER'].nunique()
print(f'The number of unique encounters is {unique_encounters}')
```

```python
# Let's read the conditions file
conditions_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\conditions.csv')
```

```python
conditions_df.head()
```

```python
# Remove columns that we don't need
conditions_df = conditions_df.drop(columns=['START', 'STOP', 'CODE'])
```

```python
# Change the name of DESCRIPTION to CONDITION_DESCRIPTION
conditions_df = conditions_df.rename(columns={'DESCRIPTION': 'CONDITION_DESCRIPTION'})

# Merge the conditions_df with the final_df
final_df = pd.merge(final_df, conditions_df, on=['PATIENT', 'ENCOUNTER'])

final_df.head()
```

```python
# Create a new column that is a list of all the conditions for each unique encounter
final_df['CONDITIONS'] = final_df.groupby('ENCOUNTER')['CONDITION_DESCRIPTION'].transform(lambda x: ','.join(x))

final_df.head()
```

```python
# Number of unique encounters
unique_encounters = final_df['ENCOUNTER'].nunique()
print(f'The number of unique encounters is {unique_encounters}')

# Size of the final_df
print(f'The size of the final_df is {final_df.shape}')
```

```python
# Remove duplicate encounters
final_df = final_df.drop_duplicates(subset='ENCOUNTER')

# Remove the condition_description column
final_df = final_df.drop(columns=['CONDITION_DESCRIPTION'])

final_df.head()

```

```python
# Size of the final_df
print(f'The size of the final_df is {final_df.shape}')
```

```python
# Create a new column called AGE

import datetime

def calculate_age(birthdate, start):
    birthdate = datetime.datetime.strptime(birthdate, '%Y-%m-%d')
    start = datetime.datetime.strptime(start, '%Y-%m-%dT%H:%M:%SZ')
    age = start.year - birthdate.year - ((start.month, start.day) < (birthdate.month, birthdate.day))
    return age

final_df['AGE'] = final_df.apply(lambda x: calculate_age(x['BIRTHDATE'], x['START']), axis=1)

final_df.head()
```

```python
# Create a new column for death event
final_df['DEATH_EVENT'] = final_df['DEATHDATE'].notnull()

final_df.head()
```

```python
# Convert Systolic Blood Pressure and Diastolic Blood Pressure to float
final_df['Systolic Blood Pressure'] = final_df['Systolic Blood Pressure'].astype(float)
final_df['Diastolic Blood Pressure'] = final_df['Diastolic Blood Pressure'].astype(float)

# Create a new column for Mean Arterial Pressure (MAP)
final_df['MAP'] = final_df['Diastolic Blood Pressure'] + 1/3 * (final_df['Systolic Blood Pressure'] - final_df['Diastolic Blood Pressure'])

final_df.head()
```

```python
# Drop Sysolic Blood Pressure and Diastolic Blood Pressure
final_df = final_df.drop(columns=['Systolic Blood Pressure', 'Diastolic Blood Pressure'])
```

```python
#Let's read the medications file
medications_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\medications.csv')
```

```python
medications_df.head()
```

```python
# Drop the columns we don't need
medications_df = medications_df.drop(columns=['START', 'STOP', 'PAYER', 'CODE', 'BASE_COST', 'PAYER_COVERAGE', 'DISPENSES', 'TOTALCOST', 'REASONCODE', 'REASONDESCRIPTION'])

medications_df.head()
```

```python
# merge the medications_df with the final_df
final_df = pd.merge(final_df, medications_df, on=['PATIENT', 'ENCOUNTER'])

final_df.head()
```

```python
# Create a new column that is a list of all the medications for each unique encounter
final_df['MEDICATIONS'] = final_df.groupby('ENCOUNTER')['DESCRIPTION'].transform(lambda x: ','.join(x))

# Drop the description column
final_df = final_df.drop(columns=['DESCRIPTION'])

final_df.head()
```

```python
# Drop duplicate encounters
final_df = final_df.drop_duplicates(subset='ENCOUNTER')

final_df.head()
```

```python
# Number of unique encounters
unique_encounters = final_df['ENCOUNTER'].nunique()
print(f'The number of unique encounters is {unique_encounters}')
```

```python
# Let's read the procedures file
procedures_df = pd.read_csv(r'10k_synthea_covid19_csv\10k_synthea_covid19_csv\procedures.csv')
```

```python
procedures_df.head()
```

```python
# Let's drop the columns we don't need
procedures_df = procedures_df.drop(columns=['DATE', 'CODE', 'BASE_COST', 'REASONCODE', 'REASONDESCRIPTION'])

procedures_df.head()
```

```python
# Merge the procedures_df with the final_df
final_df = pd.merge(final_df, procedures_df, on=['PATIENT', 'ENCOUNTER'])

final_df.head()
```

```python
# Delete duplicate encounters + procedures
final_df = final_df.drop_duplicates(subset=['ENCOUNTER', 'DESCRIPTION'])

final_df.head()
```

```python
# Let's create a new column that is a list of all the procedures for each unique encounter
final_df['PROCEDURES'] = final_df.groupby('ENCOUNTER')['DESCRIPTION'].transform(lambda x: ','.join(x))

# Drop the description column
final_df = final_df.drop(columns=['DESCRIPTION'])

final_df.head()
```

```python
# Drop duplicate encounters
final_df = final_df.drop_duplicates(subset='ENCOUNTER')

final_df.head()
```

```python
# Number of unique encounters
unique_encounters = final_df['ENCOUNTER'].nunique()
print(f'The number of unique encounters is {unique_encounters}')

# Shape of the final_df
print(f'The size of the final_df is {final_df.shape}')
```

```python
# Remove columns that we don't need
final_df = final_df.drop(columns=['ENCOUNTER', 'START', 'STOP', 'PATIENT', 'ENCOUNTER_DESCRIPTION', 'REASONDESCRIPTION', 'BIRTHDATE', 'DEATHDATE', 'MARITAL'])

final_df.head()
```

```python
# Shape of the Final DataFrame
final_df.shape
```

```python
# How many death events are in the final_df
deaths = final_df['DEATH_EVENT'].value_counts()
print('The number of death events is: ', deaths[True])

# What percentag of the final_df are death events
print('The percentage of death events is: ', f'{(deaths[True] / final_df.shape[0] * 100):.3f}%')
```
